```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
# Load data
df = pd.read_csv('Cars93.csv')
df.head()
#df.info()
#df.describe()
#df.shape
#df['Manufacturer'].unique()
```

### 1. Boxplot

```python
# Boxplot for revs per mile

#Filter data for the specified manufacturers
manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
filtered_df = df[df['Manufacturer'].isin(manufacturers)]

#Create a boxplot
plt.figure(figsize=(10, 6))
plt.title('Revs Per Mile by Manufacturer')
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=filtered_df)
plt.xticks(rotation=45)
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Toyota has got an outlier which looks like a mistake in the capturing of the data. So i would say, Suzuki has got the highest revs per mile


### 2. Histogram

```python
# Histogram of MPG

plt.figure(figsize=(12, 6))

#The first figure

plt.subplot(1, 2, 1)
plt.hist(df['MPG.highway'], bins=15, color='g', edgecolor='k')
plt.title('MPG on Highway')
plt.xlabel('MPG')
plt.ylabel('Frequency')

#The second figure
plt.subplot(1, 2, 2)
plt.hist(df['MPG.city'], bins=15, color='r', edgecolor='k')
plt.title('MPG on the City')
plt.xlabel('MPG')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?


**Histogram answer:**
It is efficient to drive on the city actually

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure(figsize=(10, 6))

sns.lineplot(x='Wheelbase', y='Turn.circle', data=df)
plt.title('Line Graph: Wheelbase vs Turning Circle')
plt.xlabel('Wheelbase')
plt.ylabel('Turning Circle')
plt.grid(True)
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
As the tuning circle increases, so does the wheelbase


### 4. Barplot

```python
# Barplot for mean horsepower
plt.figure(figsize=(10, 6))

sns.barplot(x='Type', y='Horsepower', data=df)
plt.title('Horsepower by Car Type')
plt.xlabel('Car Type')
plt.ylabel('Horsepower')
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?




**Barplot answer:**
Yes, a large car has got a high horsepower.
